```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
import pandas as pd
import numpy as np
```

```python
dfCancer=pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
dfCancer.head()
```

```python
data['target']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtes,ytrain,ytes=train_test_split(dfCancer,data['target'],random_state=101)
```

```python
import tensorflow as tf
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtes=scaler.fit_transform(xtes)
```

```python
model=tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
r=model.fit(xtrain,ytrain,validation_data=(xtes,ytes),epochs=100)
print('Train score:',model.evaluate(xtrain,ytrain))
print('Test Score:',model.evaluate(xtes,ytes))
```

```python
import matplotlib.pyplot as plt
plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val loss')
plt.legend()
plt.show()
```

```python
plt.plot(r.history['accuracy'],label='accuracy')
plt.plot(r.history['val_accuracy'],label='val accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,accuracy_score,auc,log_loss,roc_auc_score,f1_score
```

```python
result=model.predict(xtes)
prediksi=[]
for item in result:
    if item>0.5:
        prediksi.append(1)
    else:
        prediksi.append(0)
print(classification_report(prediksi,ytes))
```
